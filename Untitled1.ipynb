```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import KFold
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from keras.optimizers import Adam
from keras.utils import np_utils

# Assuming you have image data 'X' and corresponding labels 'y'
# Replace 'X' and 'y' with your actual image data and labels

# Assuming 'X' is a 4D array with shape (num_samples, height, width, channels)
# and 'y' is a 1D array with shape (num_samples,) containing class labels

# Convert labels to one-hot encoded format
num_classes = len(np.unique(y))
y_encoded = np_utils.to_categorical(y, num_classes)

# Define the neural network architecture
def create_model():
    model = Sequential()
    model.add(Flatten(input_shape=(height, width, channels)))
    model.add(Dense(64))
    model.add(Activation('relu'))
    model.add(Dense(num_classes))
    model.add(Activation('softmax'))
    return model

# Define the parameters for K-Fold cross-validation
num_folds = 5
batch_size = 32
epochs = 10

# Initialize lists to store accuracy results for each fold
accuracies = ['218, 197, 254, 254, 253, 254, 254, 253, 254, 254','151, 219, 164, 165, 213, 204, 211, 205, 171, 174']

# Create K-Fold cross-validator
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Perform K-Fold cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y_encoded[train_index], y_encoded[test_index]

    # Create and compile the model
    model = create_model()
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, verbose=1)

    # Evaluate accuracy on the test set
    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    accuracies.append(acc)

# Print the mean and standard deviation of accuracy across folds
print("Mean Accuracy:", np.mean(accuracies))
print("Standard Deviation:", np.std(accuracies))
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define data paths
train_dir = '/content/drive/MyDrive/train_images'
test_dir = '/content/drive/MyDrive/test_images'

# Define image parameters
img_width, img_height = 128, 128
input_shape = (img_width, img_height, 3)
batch_size = 32
epochs = 10

# Data augmentation for training images
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1.0/255.0)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary'
)

# Create a CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=test_generator,
    validation_steps=test_generator.samples // batch_size
)

# Plot the training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained VGG16 model
model = tf.keras.applications.VGG16(weights='imagenet', include_top=True)

# Function to preprocess the image for the VGG16 model
def preprocess_image(img):
    img = cv2.resize(img, (224, 224))
    img = tf.keras.applications.vgg16.preprocess_input(img)
    img = np.expand_dims(img, axis=0)
    return img

# Function to generate the heatmap using Class Activation Map (CAM)
def generate_heatmap(model, img):
    img = preprocess_image(img)

    # Get the last convolutional layer and the output layer of the model
    last_conv_layer = model.get_layer('block5_conv3')
    classifier_layer = model.get_layer('predictions')

    # Create a new model that outputs both the last conv layer and the classifier layer
    heatmap_model = tf.keras.models.Model([model.inputs], [last_conv_layer.output, classifier_layer.output])

    with tf.GradientTape() as tape:
        conv_output, predictions = heatmap_model(img)
        class_index = np.argmax(predictions[0])
        class_output = predictions[:, class_index]

    grads = tape.gradient(class_output, conv_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    heatmap = tf.reduce_mean(tf.multiply(pooled_grads, conv_output), axis=-1)
    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap)

    heatmap = cv2.resize(heatmap, (img.shape[2], img.shape[1]))  # Resize heatmap to match original image size
    heatmap = np.uint8(255 * heatmap)

    return heatmap

# Example usage:
image_path = '/content/drive/MyDrive/test_images/200004.jpg'  # Replace with the path to your paddy leaf image

# Read the image using OpenCV
image = cv2.imread(image_path)

# Calculate the heatmap from the image
heatmap_image = generate_heatmap(model, image)

# Display the heatmap image using matplotlib
plt.imshow(heatmap_image, cmap='jet')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def sir_model(S, I, R, beta, gamma, N):
    dS = -beta * S * I / N
    dI = beta * S * I / N - gamma * I
    dR = gamma * I
    return dS, dI, dR
# Initial conditions
S0 = 990
I0 = 10
R0 = 0

# Total population
N = S0 + I0 + R0

# Model parameters
beta = 0.3  # Transmission rate
gamma = 0.1  # Recovery rate
# Time grid for simulation
num_days = 100
time_points = np.linspace(0, num_days, num_days + 1)
# Lists to store results
S = [S0]
I = [I0]
R = [R0]

# Euler's method for numerical integration
for day in range(1, num_days + 1):
    dS, dI, dR = sir_model(S[-1], I[-1], R[-1], beta, gamma, N)
    S_new = S[-1] + dS
    I_new = I[-1] + dI
    R_new = R[-1] + dR
    S.append(S_new)
    I.append(I_new)
    R.append(R_new)
# Plot the results
plt.plot(time_points, S, label='Susceptible')
plt.plot(time_points, I, label='Infectious')
plt.plot(time_points, R, label='Recovered')
plt.xlabel('Time (days)')
plt.ylabel('Number of Individuals')
plt.title('SIR Model Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def find_tangent_line(parabola_equation, x_value):
    # Define the variable and the parabola equation using SymPy
    x = sp.Symbol('x')
    parabola = sp.sympify(parabola_equation)

    # Calculate the derivative of the parabola
    derivative = sp.diff(parabola, x)

    # Calculate the y-value of the point on the parabola
    y_value = parabola.subs(x, x_value)

    # Calculate the slope of the tangent line
    slope = derivative.subs(x, x_value)

    # Calculate the y-intercept of the tangent line
    y_intercept = y_value - slope * x_value

    # Define the equation of the tangent line
    tangent_equation = slope * x + y_intercept

    return tangent_equation

def main():
    # Define the equation of the parabola
    parabola_equation = 'x**2'  # Change this to your desired parabola equation

    # Point of tangency (x-value)
    x_value = 2  # Change this to your desired x-value for the tangent point

    # Find the equation of the tangent line
    tangent_equation = find_tangent_line(parabola_equation, x_value)

    # Convert the tangent equation to a lambda function for plotting
    x = np.linspace(x_value - 5, x_value + 5, 100)
    tangent_function = sp.lambdify('x', tangent_equation, 'numpy')
    tangent_y = tangent_function(x)

    # Plot the parabola and the tangent line
    parabola_function = sp.lambdify('x', parabola_equation, 'numpy')
    parabola_y = parabola_function(x)

    plt.plot(x, parabola_y, label='Parabola')
    plt.plot(x, tangent_y, label='Tangent Line')
    plt.scatter(x_value, tangent_function(x_value), color='red', label='Tangent Point')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Tangent Line to the Parabola')
    plt.show()

if __name__ == '__main__':
    main()
```

```python
import numpy as np
import tensorflow as tf
from keras.layers import Layer, Activation
from keras import backend as K
from keras.models import Sequential
from keras.layers import Dense

# Define the custom Swish activation function with user input for beta
class Swish(Layer):
    def __init__(self, beta, **kwargs):
        super(Swish, self).__init__(**kwargs)
        self.beta = beta

    def call(self, inputs):
        return inputs * K.sigmoid(self.beta * inputs)

    def get_config(self):
        config = super(Swish, self).get_config()
        config.update({'beta': self.beta})
        return config

# User input for beta
beta = float(input("218, 197, 254, 254, 253, 254, 254, 253, 254, 254 "))

# Create the custom Swish activation function with user input for beta
swish_custom = Swish(beta)

# Create a simple Keras model with Swish activation function
model = Sequential()
model.add(Dense(64, input_shape=(10,)))
model.add(swish_custom)  # Using the custom Swish activation function
model.add(Dense(32))
model.add(swish_custom)  # Using the custom Swish activation function
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Generate random input data for demonstration
X = np.random.rand(100, 10)
y = np.random.randint(2, size=(100, 1))

# Train the model
model.fit(X, y, epochs=10, batch_size=16, verbose=1)
```
